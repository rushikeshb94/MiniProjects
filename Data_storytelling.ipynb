## Which category has the maximum and minimum # of drinks

```python
sns.set_theme(style="whitegrid")
ax = sns.countplot(y="Beverage_category", data=df,  order = df['Beverage_category'].value_counts().index)

```

The Classic Espresso Drinks have the maximum number of drinks on the menu while Coffee category has the least

## Which are top 5 drinks in terms of Calorie count in the menu

```python
plt.figure(figsize=(10, 8))
df2 =df[['Beverage', 'Calories', 'Beverage_prep']]
df2 =df2.sort_values(['Calories'], ascending  = False)

df2 = df2.reset_index(drop=True)
ax = sns.barplot(x='Beverage', y='Calories', hue='Beverage_prep', data = df2.head()) 
plt.xticks(rotation=40, ha='right')
plt.show()
```

The top five drinks in terms of calories are white chocalate mocha with 2% milk, Java Chip with whole mil, white chocalate mocha with soymilk, white chacoalate mocha with non fat milk, adn tazo green tea latte with 2% milk in that order. It is interesting to note that the white chocalate mocha with 2% milk has a calorie calue of 510 while the same drink prepared with Soymilk has a calorie of 450.

## Which are top 5 drinks in terms of Sugars (g) count in the menu

```python
plt.figure(figsize=(10, 8))
df3 =df[['Beverage', ' Sugars (g)', 'Beverage_prep']]
df3 =df3.sort_values([' Sugars (g)'], ascending  = False)

df3 = df3.reset_index(drop=True)
ax = sns.barplot(x='Beverage', y=' Sugars (g)', hue='Beverage_prep', data = df3.head()) 
plt.xticks(rotation=40, ha='right')
plt.show()
```

The top five drinks in terms of sugar content are Java chip with whole milk, Java Chip with venti non-fat milk, Caramel Apple spice, Java chip wiht soy milk, and Caramel with whole milk in that order.Java chip drink has a high sugar content irrespective of the preparation type.

## Correlation between variable

```python
# Calories vs Sugars (g)
plt.figure(figsize=(10, 10))
ax = sns.scatterplot(data=df, x="Calories", y=" Sugars (g)", hue="Beverage_category")
```

```python
#correlation plot between different variables 
corr = df.corr(method='kendall')
plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True, cmap="Greens")
```
